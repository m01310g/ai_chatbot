```python
import pandas as pd

file3 = pd.read_csv('C:/Users/hello/Downloads/07_24_04_P_CSV/fulldata_07_24_04_P_일반음식점.csv', encoding='cp949')
file3 = file3[['영업상태명', '사업장명', '소재지전체주소', '좌표정보(x)', '좌표정보(y)']]
file3.drop(file3[file3['영업상태명'] == '폐업'].index, inplace=True)
file3

file3['좌표정보(x)'] = pd.to_numeric(file3['좌표정보(x)'], errors='coerce')
file3['좌표정보(y)'] = pd.to_numeric(file3['좌표정보(y)'], errors='coerce')
file3 = file3.dropna()
file3.index = range(len(file3))
file3
```

```python
import pyproj
import numpy as np
import folium

def project_array(coord, p1_type, p2_type):
    """
    좌표계 변환 함수
    - coord: x, y 좌표 정보가 담긴 NumPy Array
    - p1_type: 입력 좌표계 정보 ex) epsg:5179
    - p2_type: 출력 좌표계 정보 ex) epsg:4326
    """
    p1 = pyproj.Proj(init=p1_type)
    p2 = pyproj.Proj(init=p2_type)
    fx, fy = pyproj.transform(p1, p2, coord[:, 0], coord[:, 1])

    return np.dstack([fx, fy])[0]
```

```python
coord = np.array(file3[['좌표정보(x)', '좌표정보(y)']])
p1_type = "epsg:2097"
p2_type = "epsg:4326"

result = project_array(coord, p1_type, p2_type)
result
```

```python
file3['경도'] = result[:, 0]
file3['위도'] = result[:, 1]
file3
```

```python
file3.drop('좌표정보(x)', axis=1, inplace=True)
file3.drop('좌표정보(y)', axis=1, inplace=True)
```
